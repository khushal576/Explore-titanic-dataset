### 1.) Missing Value:

### Types of Missing Values

Understanding the nature of missing data is crucial for selecting appropriate imputation techniques. Missing data can be classified into three main categories:

#### 1. Missing Completely at Random (MCAR)

**Definition**: Missing values are considered MCAR when the likelihood of data being missing is independent of both observed and unobserved data. In other words, there is no pattern to the missingness.

**Example**: A survey respondent accidentally skips a question due to a printing error in the questionnaire.

**Implications**: Data analysis methods that assume MCAR are generally unbiased. Simple imputation techniques like mean, median, or listwise deletion can be appropriate if the proportion of missing data is small.

#### 2. Missing at Random (MAR)

**Definition**: Missing values are considered MAR when the probability of missingness is related to observed data but not the missing data itself. In other words, the missingness can be explained by other variables in the dataset.

**Example**: In a medical study, older patients might be less likely to provide their income information, but the missingness is related to age, which is observed.

**Implications**: More sophisticated imputation methods like multiple imputation or model-based approaches are required, as the missingness is not completely random but can be accounted for using other observed variables.

#### 3. Missing Not at Random (MNAR)

**Definition**: Missing values are considered MNAR when the probability of missingness is related to the missing data itself. This means that there is a systematic difference between the missing and observed values.

**Example**: People with higher incomes might be less likely to report their income, and the missingness directly relates to the income values themselves.

**Implications**: Handling MNAR data is challenging because the missingness mechanism is related to the unobserved data. Specialized techniques or assumptions are often required, and it may be necessary to collect additional data or conduct sensitivity analyses.

### Approach we are following is:
* **Group-Based Imputation with Titles:** This approach combines domain knowledge and statistical imputation, using the median age within groups defined by passenger class and title. This method is effective because it considers the socio-economic status and typical age ranges associated with different titles, leading to more accurate imputation.<br/>
   **1. Extract Titles from Names:** Titles (e.g., Mr., Mrs., Miss, Master) provide valuable information about the passengers' demographics and can be used to group passengers.

```python
# Extract titles from names column
df['Title'] = df['Name'].str.extract('([A-Za-z]+)\.', expand=False)

# Standardize titles
# common titles
df['Title'] = df['Title'].replace(['Mlle', 'Ms', 'Mme'], ['Miss', 'Miss', 'Mrs'])
# Rare title
df['Title'] = df['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 
                                       'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
```

**2. Group by Pclass and Title and Impute Missing Ages:** Calculate the median age for each group defined by Pclass and Title, then fill the missing values in the Age column with these medians.

```python
# Calculate median ages for each Pclass and Title group
median_ages = df.groupby(['Pclass', 'Title'])['Age'].median()

# Define a function to fill missing ages
def impute_age(row):
    if pd.isnull(row['Age']):
        return median_ages[row['Pclass'], row['Title']]
    else:
        return row['Age']

# Apply the function to fill missing ages
df['Age'] = df.apply(impute_age, axis=1)
```

<div class="alert alert-block alert-info">
<b>Note:</b> We chose the median for imputing missing values in the "Age" column because it is robust to outliers and better represents the central tendency of the typically skewed age distribution in the Titanic dataset. 
</div>

## Preprocessing of data:

###  Split new_df to training and testing set.
* train_test_split(): This function from scikit-learn splits the data into random train and test subsets. It takes the features (X) and the target variable (y) as inputs.
* test_size=0.2: Specifies that 20% of the data will be allocated to the test set, while the remaining 80% will be assigned to the training set.
* random_state: A random seed used to ensure reproducibility of the split. Using the same random state ensures that the split will be the same every time you run the code.
* The resulting datasets are X_train, X_test, y_train, and y_test, which contain the features and target variable for the training and test sets, respectively.

```python
from sklearn.model_selection import train_test_split
# Split into X and y
X = new_df.drop("Survived", axis=1)
y = new_df["Survived"]

# Split data into train and test
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2)

# Print the shapes of the resulting datasets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

### Feature Transformation:

#### 1.) Encoding Categorical Variables:
Categorical variables need to be encoded into numerical format for machine learning models to process them. Common encoding methods include:

* **One-Hot Encoding:** Creates binary columns for each category, indicating its presence or absence.
* **Label Encoding:** Assigns a unique numerical label to each category.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Define the columns to encode
columns_to_encode = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked', 'Title', 'FamilySize']

# Create the OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)

# Define the transformer
transformer = ColumnTransformer([
    ('one_hot_encoding', encoder, columns_to_encode)
], remainder='passthrough')

# Fit the transformer on the training data
transformer.fit(X_train)

# Transform both the training and test data
train_encoded_data = transformer.transform(X_train)
test_encoded_data = transformer.transform(X_test)

# Get the feature names for the one-hot encoded columns
encoded_columns = transformer.named_transformers_['one_hot_encoding'].get_feature_names_out(columns_to_encode)

# Combine encoded column names with the names of the remainder columns
remainder_columns = X_train.columns.difference(columns_to_encode)
column_names = list(encoded_columns) + list(remainder_columns)

# Create DataFrames with encoded data and column names
train_encoded_df = pd.DataFrame(train_encoded_data, columns=column_names)
test_encoded_df = pd.DataFrame(test_encoded_data, columns=column_names)
```

# Scaling and Its Types

Scaling is a data preprocessing technique used to standardize the range of independent variables or features of data. In machine learning, scaling ensures that features contribute equally to the result, prevents features with larger ranges from dominating the learning process, and improves the convergence speed of some algorithms.

## Types of Scaling

### 1. Min-Max Scaling (Normalization)
  Min-Max Scaling scales the data to a fixed range, typically [0, 1] or [-1, 1]. It is useful when the data needs to be in a specific range.<div>
  $$
  X_{\text{scaled}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}
  $$
</div>

- **Use Cases**: When the distribution of data is not Gaussian (not bell-shaped) or when using algorithms that do not assume any distribution of data (e.g., K-Nearest Neighbors).

### 2. Standardization (Z-score Normalization)
Standardization scales the data to have a mean of 0 and a standard deviation of 1. This is useful when the data follows a Gaussian distribution.
  <div>
  $$
  X_{\text{scaled}} = \frac{X - \mu}{\sigma}
  $$
</div>

- **Use Cases**: When the data is normally distributed or when using algorithms like linear regression, logistic regression, or neural networks.

### 3. Robust Scaling
Robust Scaling uses the median and interquartile range (IQR) for scaling, making it robust to outliers.

  <div>
  \[
  X_{\text{scaled}} = \frac{X - \text{median}}{\text{IQR}}
  \]
</div>

- **Use Cases**: When the data contains many outliers, as it reduces the influence of outliers.

### 4. MaxAbs Scaling
MaxAbs Scaling scales each feature by its maximum absolute value. It preserves the sparsity of the data.

  <div>
  \[
  X_{\text{scaled}} = \frac{X}{|X_{\text{max}}|}
  \]
</div>

- **Use Cases**: When dealing with data that contains both positive and negative values or when working with sparse data (e.g., text data).

### 5. Quantile Transformation (Rank Scaling)
Quantile Transformation transforms features to follow a uniform or normal distribution. This method maps the data to a desired distribution based on the cumulative distribution function (CDF).
- **Use Cases**: When a feature does not follow a Gaussian distribution and you want to transform it to be normally distributed, which can be beneficial for some models.

### 6. Log Transformation
Log Transformation applies the natural logarithm to each feature. This can help reduce skewness and make the data more normally distributed.

  <div>
  \[
  X_{\text{scaled}} = \log(X + 1)
  \]
</div>

- **Use Cases**: When the data has a right-skewed distribution (many small values and few large values).

<div class="alert alert-block alert-info">
When most of your values are 0 or 1, scaling might not be necessary, especially for algorithms that are not sensitive to the scale of the data (such as tree-based methods). However, for algorithms that are sensitive to the scale, like linear regression, logistic regression, or neural networks, scaling can still be beneficial.</div>

# Importance of Scaling in Different Algorithms

|------------------------------------------------------|-------------------------------------------------------------|

- **Algorithms where Scaling is Important**: These algorithms are sensitive to the magnitude of the features and scaling helps them converge faster and perform better.
- **Algorithms where Scaling is Not Typically Important**: These algorithms are generally invariant to feature scaling as they are based on decision trees or rules that do not rely on the distance or magnitude of the features.
